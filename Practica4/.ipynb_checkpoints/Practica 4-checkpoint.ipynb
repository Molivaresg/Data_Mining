```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('FashionDataPoints.csv')
```

```python
df.head()
```

```python
df.columns
df.info()
df.isnull().sum()
```

```python
# Remove empty columns
df_cleaned = df.dropna(axis=1, how='all')
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.dropna(axis=0, how='any')
df_cleaned.shape
```

```python
# Changes names
new_names = [
    'Timestamp', 'Age_Group', 'Gender', 'Profession',
    'Daily_Outfit', 'Color_Palette', 'Functionality_vs_Aesthetics',
    'Wardrobe_Style', 'Shopping_Frequency', 'Purchase_Influence',
    'Shopping_Place', 'Formal_Events', 'Style_Experimentation',
    'Footwear_Type', 'Lifestyle_Activity', 'Comfort_Importance',
    'Timeless_vs_Trendy', 'Style_Personality'
]
```

```python
df_cleaned.columns = new_names
df_cleaned.head()
```

```python
# Summary statistics for numerical columns
df_cleaned.describe()
```

```python
# Count unique values in categorical columns
df_cleaned['Gender'].value_counts()
df_cleaned['Age_Group'].value_counts()
df_cleaned['Daily_Outfit'].value_counts()
```

```python
# Most popular outfit style
df_cleaned['Daily_Outfit'].value_counts()
```

```python
# chi-square test to see if there is a significant relationship between gender and preference for functionality or aesthetics

table = pd.crosstab(df_cleaned['Gender'], df_cleaned['Functionality_vs_Aesthetics'])

chi2, p, dof, expected = stats.chi2_contingency(table)

print("Chi-square =", chi2)
print("P-value =", p)

if p < 0.05:
    print("We reject the null hypothesis → There is a relationship between gender and preference for functionality/aesthetics.")
else:
    print("We cannot reject the null hypothesis → There is no evidence of a relationship.")
```

```python
comfort_map = {
    '- Not important at all': 1,
    '- Somewhat unimportant': 2,
    '- Neutral': 3,
    '- Somewhat important': 4,
    '- Extremely important': 5
}
```

```python
df_cleaned['Comfort_Importance_Num'] = df_cleaned['Comfort_Importance'].map(comfort_map)
```

```python
females = df_cleaned[df_cleaned['Gender'] == 'Female']['Comfort_Importance_Num'].dropna()
males = df_cleaned[df_cleaned['Gender'] == 'Male']['Comfort_Importance_Num'].dropna()
```

```python
from scipy.stats import ttest_ind

#Student's t-test to compare if there is a significant difference

t_stat, p_val = ttest_ind(females, males)

print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("Significant difference between males and females!")
else:
    print("No significant difference.")
```
