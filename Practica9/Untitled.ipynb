In this analysis, I explore a set of Amazon product reviews to extract key insights from customer feedback. Using sentiment analysis, I classify the emotional tone of each review, while topic modeling helps reveal the main themes discussed. The goal is to interpret the results, evaluate the relevance of the extracted information, and assess its applicability in real-world contexts, such as improving products or understanding customer needs.

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation


df = pd.read_csv('reviews.csv')
print(df.head())

```

```python
#Cleaning the Text Data
def clean_text(text):
    text = str(text).lower()
    return ''.join([char for char in text if char.isalnum() or char.isspace()])

df['cleaned_text'] = df['reviews.text'].apply(clean_text)
```

```python
#Sentiment Analysis with TextBlob
#Returns a sentiment score between -1 and 1 (negative to positive)
#Classifies each review as Positive, Negative, or Neutral based on the polarity score.
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

def get_sentiment(polarity):
    if polarity > 0:
        return 'Positivo'
    elif polarity < 0:
        return 'Negativo'
    else:
        return 'Neutro'

df['polarity'] = df['cleaned_text'].apply(get_polarity)
df['sentiment'] = df['polarity'].apply(get_sentiment)
```

```python
#Converts the cleaned text into numerical representations (TF-IDF).
#Removes words that appear in more than 90% of the documents
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.9)
tfidf = tfidf_vectorizer.fit_transform(df['cleaned_text'])
```

```python
#Applying LDA (Latent Dirichlet Allocation) for Topic Modeling
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(tfidf)

# Display the Top Words of Each Topic
n_words = 10
for index, topic in enumerate(lda.components_):
    print(f"Topic #{index + 1}:")
    top_words = [tfidf_vectorizer.get_feature_names_out()[i] for i in topic.argsort()[:-n_words - 1:-1]]
    print(" ".join(top_words))
```

```python
#Assign the Dominant Topic to Each Review
topic_distribution = lda.transform(tfidf)
df['dominant_topic'] = topic_distribution.argmax(axis=1)
```

```python
# Cross Sentiment with Topic
topic_sentiment = df.groupby('dominant_topic')['sentiment'].value_counts(normalize=True).unstack()
print(topic_sentiment)
```

```python
df['sentiment'].value_counts().plot(kind='bar', color=['green', 'gray', 'red'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()

```

```python
df['dominant_topic'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Reviews per Topic')
plt.xlabel('Topic')
plt.ylabel('Number of Reviews')
plt.show()

```

```python
#Generate Word Clouds for Each Topic

from wordcloud import WordCloud

for index, topic in enumerate(lda.components_):
    print(f"Topic #{index + 1}")
    top_words_dict = {tfidf_vectorizer.get_feature_names_out()[i]: topic[i] for i in topic.argsort()[:-11:-1]}
    wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(top_words_dict)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Topic #{index + 1}')
    plt.show()
```
